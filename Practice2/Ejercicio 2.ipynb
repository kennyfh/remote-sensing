```python
# Importación módulos
import sys
import numpy as np
np.set_printoptions(threshold=10000)
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creamos la función expan
def expan(img,cmin,cmax):
    # Copiar imagen
    resimg = img.copy()
    fila,columna = img.shape
    
    for i in range(fila):
        for j in range(columna):
            pixel = resimg[i,j]
            if pixel <= cmin:
                resimg[i,j] = 0
            elif pixel >= cmax:
                resimg[i,j] = 255
            else:
                resimg[i,j] = np.round(0 + ((pixel-cmin)/(cmax-cmin))*(255-0))
    return resimg
```

```python
ruta =  '../imgpack/Monobanda/2_Corsica.jpg'
img = cv.imread(ruta,cv.IMREAD_GRAYSCALE) #Transformamos nuestra imagen a escala de grises
# Aplica la expansión lineal
img_exp = expan(img,10,120)

#Dibuja el histograma entre 0 y 254
plt.hist(img_exp.ravel(),255,[0,254],width=1)
plt.title('Hisograma de la imagen usando expansión lineal')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10)) # Creamos nueva figura
    
#Mostramos la imagen como el histograma
fig.add_subplot(221)
plt.title('Imagen sin realizar expansión lineal')
plt.imshow(img,cmap='gray')
fig.add_subplot(222)
plt.title('Imagen usando expansión lineal')
plt.imshow(img_exp,cmap='gray')
plt.show()
```
