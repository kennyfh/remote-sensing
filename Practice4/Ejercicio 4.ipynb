```python
# Importación módulos
import sys
import numpy as np
np.set_printoptions(threshold=10000)
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Función máscara
def mascara(img,mask):
    resimg = img.copy()
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            resimg[x,y] = resimg[x,y]*mask[x,y]
    return resimg

#Función calcula máscara
def calculaMask(img,NDmin,NDmax):
    f,c = img.shape
    mask = np.zeros((f,c),dtype=np.int16)
    for x in range(f):
        for y in range(c):
            pix = img[x,y]
            mask[x,y] = 1 if pix > NDmin and pix < NDmax else 0
            
    return mask
```

```python
ruta =  '../imgpack/Filomena.jpg' # Consideramos que la imagen es monobanda
image = cv.imread(ruta,cv.IMREAD_GRAYSCALE) #Transformamos nuestra imagen a escala de grises
```

```python
mask_nubes= calculaMask(image,150,200)
```

```python
mask_nieves = calculaMask(image,220,255)
```

```python
#Imagen de ejemplo
fig = plt.figure(figsize=(15, 15)) # Creamos nueva figura

fig.add_subplot(221)
plt.imshow(image,cmap='gray')
plt.show()
```

```python
msk_img = mascara(image,mask_nubes)
fig = plt.figure(figsize=(15, 15)) # Creamos nueva figura
fig.add_subplot(221)
plt.title('Resultado aplicar máscara')
plt.imshow(msk_img,cmap='gray')
plt.show()
```

```python
msk_img2 = mascara(image,mask_nieves)
fig = plt.figure(figsize=(15, 15)) # Creamos nueva figura
fig.add_subplot(221)
plt.title('Resultado aplicar máscara')
plt.imshow(msk_img2,cmap='gray')
plt.show()
```
