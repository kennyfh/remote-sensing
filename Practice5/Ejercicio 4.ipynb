```python
# Importación módulos
import sys
import numpy as np
np.set_printoptions(threshold=10000)
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ruta1 =  'ejercicio3.png'
img= cv.imread(ruta1,cv.IMREAD_GRAYSCALE)
```

```python
#Función densi
def densi(img):
    
    densi_res = np.zeros((img.shape[0],img.shape[1],3),np.uint8)
    colores= [(255,0,0),(65,105,225),(255,192,203),(255,215,0)]
    # Rojo, #azul real, rosado,oro
    #128-192
    rangos= [(128,140),(141,150),(151,160),(161,192)]
    
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            pix = img[x,y]
            if pix >=rangos[0][0] and pix<rangos[0][1]:
                densi_res[x,y]= colores[0]
            elif pix >=rangos[1][0] and pix<rangos[1][1]:
                densi_res[x,y]= colores[1]
            elif pix >=rangos[2][0] and pix<rangos[2][1]:
                densi_res[x,y]= colores[2]
            elif pix >=rangos[3][0] and pix<rangos[3][1]:
                densi_res[x,y]= colores[3]
            else:
                densi_res[x,y]=pix
    return densi_res
```

```python
img_exp= densi(img)
```

```python
fig = plt.figure(figsize=(15, 15)) # Creamos nueva figura
fig.add_subplot(221)
plt.imshow(img_exp)
plt.show()
```
